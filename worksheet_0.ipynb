## Problem 4: Matrix Operations

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

### 1. Prove A · A⁻¹ = I

```python
A_inv = np.linalg.inv(A)
print(np.dot(A, A_inv))
```

### 2. Show AB ≠ BA

```python
AB = np.dot(A,B)
BA = np.dot(B,A)

print('AB =')
print(AB)

print('\nBA =')
print(BA)
```

### 3. Prove (AB)ᵀ = BᵀAᵀ

```python
left = np.transpose(np.dot(A,B))
right = np.dot(np.transpose(B), np.transpose(A))

print(left)
print(right)
```

## Solve Linear System Using NumPy

---



Equations:

2x − 3y + z = −1  
x − y + 2z = −3  
3x + y − z = 9

```python
A = np.array([
    [2,-3,1],
    [1,-1,2],
    [3,1,-1]
])

B = np.array([-1,-3,9])
```

### Solve using NumPy linear solver

```python
solution = np.linalg.solve(A,B)
print('Solution [x,y,z]:', solution)
```
